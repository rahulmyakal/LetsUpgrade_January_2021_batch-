```
 
import random
import string
choice = list(string.ascii_uppercase)
movies = ["war", "dangal","darbar","heropanti","joker"]
 
def check(qn):
    return "*" in qn
def create_question(movie):
    n = len(movie)
    letters = list(movie)
    temp = []
    for i in range(0, n):
        if letters[i] == " ":
            temp.append(" ")
        else:
            temp.append("*")
    qn = "".join(str(x) for x in temp)
    return qn
 
 
def is_present(letter, movie):
    return letter in movie
 
 
def unlock(qn, movie, letter):
    ref = list(movie)
    ql = list(qn)
    temp = []
    n = len(movie)
    for i in range(n):
        if ref[i] == letter or ref[i] == ' ':
            temp.append(ref[i])
        else:
            temp.append(ql[i])
    qn = "".join(x for x in temp)
    return qn
 
 
def play():
    picked_movie = random.choice(movies)
    picked_movie = picked_movie.upper()
    qn = create_question(picked_movie)
    print(qn)
    modified_qn = qn
    not_said = True
    while not_said:
        letter = random.choice(choice)
        choice.remove(letter)
        print("Please enter a letter :",letter)
        if is_present(letter, picked_movie):
            modified_qn = unlock(modified_qn, picked_movie, letter)
            print("Good going!",letter,"has been found in the Movie Name.")
            print()
            print(modified_qn)
            if not(check(modified_qn)):
                print("Well Done! You have guessed the Movie Name correctly!")
                not_said = False
        else:
            print(letter,"is not found in the Movie Name. Please try again!")
            print()
 
 
play()
print("\n\nTHANK YOU SO MUCH FOR TRYING THIS GAME! Comment down your favourite movie so that I can add that movie to the movie list. Have a good day :)")
```
